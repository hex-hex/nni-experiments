```python
hidden_factors = [
      4,
      8,
      16,
      32,
      64,
      128
    ]
batch = [
      64,
      128,
      256,
      512,
      1024,
      2048
    ]
regularizer = range(-8, -4, 1)
lr = range(-4, -2, 1)
```

```python
_value = []
for h in hidden_factors:
    for b in batch:
        for r in regularizer:
            for l in lr:
                _value.append({
                    'hidden_factors': h,
                    'batch': b,
                    'regularizer': r,
                    'lr': l
                })
```

```python
print(json.dumps(_value, indent=2))
```
